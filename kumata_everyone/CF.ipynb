```python
import numpy

def get_rating_error(r, p, q):
    s = 0
    for i in range(len(p)):
        s += p[i]*q[i]
    return r - s


def get_error(R, P, Q, beta):
    error = 0.0
    for i in range(len(R)):
        for j in range(len(R[i])):
            if R[i][j] == 0:
                continue
            error += pow(get_rating_error(R[i][j], P[i], Q[j]), 2)
    error += beta/2.0 * (numpy.linalg.norm(P) + numpy.linalg.norm(Q))
    return error


def matrix_factorization(R, K, steps=5000, alpha=0.0002, beta=0.02, threshold=0.001):
    P = numpy.random.rand(len(R),K).tolist()
    Q = numpy.random.rand(len(R[0]),K).tolist()
    for step in range(steps):
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] == 0:
                    continue
                err = get_rating_error(R[i][j], P[i], Q[j])
                for k in range(K):
                    P[i][k] += alpha * (2 * err * Q[j][k])
                    Q[j][k] += alpha * (2 * err * P[i][k])
        error = get_error(R, P, Q, beta)
        if error < threshold:
            break
    return P, Q


R = numpy.array([
        [5, 3, 0, 1],
        [4, 0, 0, 1],
        [1, 1, 0, 5],
        [1, 0, 0, 4],
        [0, 1, 5, 4],
        ]
    )
nP, nQ = matrix_factorization(R, 2)
nR = numpy.dot(np.array(nP), np.array(nQ).T)
nR
```
