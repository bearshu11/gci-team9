```python
import pandas as pd
dfA=pd.read_csv("train_A.tsv", delimiter='\t')
dfB=pd.read_csv("train_B.tsv", delimiter='\t')
dfC=pd.read_csv("train_C.tsv", delimiter='\t')
dfD=pd.read_csv("train_D.tsv", delimiter='\t')
df_list=[dfA,dfB,dfC,dfD]
```

```python
df =dfA.sort_values(by=["user_id","product_id","time_stamp"],ascending = [True,True,True])
df["Counts"] = df.groupby(["product_id","user_id","event_type"])["event_type"].transform('count')
a = df.duplicated(["product_id","user_id","event_type"],keep=False)
df  = df[a]
df = df.drop_duplicates(["user_id","product_id","event_type"])
df.ix[:,['user_id','product_id','event_type','Counts','time_stamp','ad']]
```
